```python
import folium as fo
import requests
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/2020_coronavirus_pandemic_in_Indonesia'

df = pd.read_html(url,header=0)
df = df[3]
df = df[['Province','Confirmed','Recovered','Deaths','Active']]
df.drop([0,35,36,37],inplace=True)
df.replace('Jakarta[b]','Jakarta', inplace=True)
df = df.reset_index()
df.drop('index',1,inplace=True)
```

```python
df
```

```python
jsonurl = 'https://raw.githubusercontent.com/LintangWisesa/Indonesia-Covid19-Maps/master/data/gps_indonesia.json'

loc = pd.read_json(jsonurl)
loc
```

```python
covidFrame = pd.concat([df,loc],axis=1)
covidFrame.drop('Provinsi',1,inplace=True)
covidFrame
```

```python
covidMap = fo.Map(
    location = [0.7893, 113.9213],
    tiles = 'OpenStreetMap', # 'Stamen Terrain',
    zoom_start = 5
)

for prov,conf,rec,death,act,lat,long,icon in covidFrame.values:
    fo.Marker(
        [lat,long],
        tooltip=prov,
        icon=fo.features.CustomIcon(icon,icon_size=(35,30)),
        popup=f"{prov}<br>Confirmed: {conf} | Recovered: {rec} | Deaths: {death} | Active: {act}"
    ).add_to(covidMap)
    
covidMap
```

```python
covidMap.save('covid19.html')
```
